```python
df_titanic = pd.read_csv("../data/titanic.csv")
df_titanic.head()
```

```python
plt.scatter(df_titanic["Age"], df_titanic["Fare"])
plt.show()
```

```python
plt.scatter(df_titanic["Age"], df_titanic["Pclass"])
plt.show()
```

```python
# Solution 1
df_titanic_male = df_titanic[df_titanic["Sex"] == 'male']
df_titanic_female = df_titanic[df_titanic["Sex"] == 'female']

plt.scatter(df_titanic_male["Age"], df_titanic_male["Survived"], c='blue', alpha=0.5)
plt.scatter(df_titanic_female["Age"], df_titanic_female["Survived"], c='red', alpha=0.5)
plt.yticks([0, 1], ['Not Survived', 'Survived'])
plt.xticks(rotation=45)
plt.legend(['Male', 'Female'])
plt.xlabel('Age')
plt.ylabel('Survived')
plt.show()
```

```python
# Solution 2
df_titanic["Sex_color"] = np.where(df_titanic["Sex"] == 'male', '#0000ff', '#ff0000')

plt.scatter(df_titanic["Age"], df_titanic["Survived"], c=df_titanic["Sex_color"], alpha=0.5)
plt.yticks([0, 1], ['Not Survived', 'Survived'])
plt.xticks(rotation=45)
plt.xlabel('Age')
plt.ylabel('Survived')
plt.show()
```

## Box Plot

```python
plt.boxplot(df_titanic[["Age", "Fare"]].dropna())
plt.xticks([1, 2], ['Age', 'Fare'])
plt.show()
```

```python
plt.boxplot([df_titanic[df_titanic["Sex"] == 'male']["Age"].dropna(),
             df_titanic[df_titanic["Sex"] == 'female']["Age"].dropna()])
plt.xticks([1, 2],['Male', 'Female'])
plt.ylabel('Age')
plt.show()
```

## Violin Plot

```python
plt.violinplot([df_titanic[df_titanic["Sex"] == 'male']["Age"].dropna(),
                df_titanic[df_titanic["Sex"] == 'female']["Age"].dropna()],
               showmedians=True,
               # showmeans=True,
               vert=True)
plt.xticks([1, 2],['Male', 'Female'])
plt.ylabel('Age')
plt.show()
```
