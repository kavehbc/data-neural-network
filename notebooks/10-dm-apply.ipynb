# Pandas Apply Function

```python
import pandas as pd
import numpy as np
import time
```

```python
df = pd.read_csv("../data/titanic.csv")
df.head()
```

## np.where()

```python
t0 = time.time()

df["Gender"] = 0
df["Gender"] = np.where(df["Sex"] == "male", 1, df["Gender"])

t1 = time.time()
print(f"Took {(t1 - t0):5f} seconds")

df.head()
```

## pd.iterrows()

```python
t0 = time.time()

df["Gender"] = 0

for index, row in df.iterrows():
    if row["Sex"] == "male":
        df.loc[index, "Gender"] = 1
    else:
        df.loc[index, "Gender"] = 0

t1 = time.time()
print(f"Took {(t1 - t0):5f} seconds")

df.head()
```

## Apply()

```python
def transform_gender(x):
    if x == "male":
        return 1
    else:
        return 0

t0 = time.time()

df["Gender"] = 0
df["Gender"] = df["Sex"].apply(transform_gender)

t1 = time.time()
print(f"Took {(t1 - t0):5f} seconds")

df.head()
```

## Apply() with Lambda

```python
t0 = time.time()

df["Gender"] = 0
df["Gender"] = df["Sex"].apply(lambda x: 1 if x == "male" else 0)

t1 = time.time()
print(f"Took {(t1 - t0):5f} seconds")

df.head()
```
