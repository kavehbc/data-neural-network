# Visualization / Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/titanic.csv")
df.head()
```

```python
df["Age"].hist(bins=80)
plt.show()
```

```python
plt.hist(df["Age"], bins=80)
plt.show()
```

```python
plt.hist(df[["Pclass", "SibSp"]], bins=8)
plt.legend(["Passenger Class", "Siblings"])
plt.show()
```

```python
df_male = df[df["Sex"] == 'male']
df_female = df[df["Sex"] == 'female']

plt.hist([df_male["Age"], df_female["Age"]], bins=30)
plt.legend(["Male", "Female"])
plt.show()
```

```python
df_class_1 = df[df["Pclass"] == 1]
df_class_2 = df[df["Pclass"] == 2]
df_class_3 = df[df["Pclass"] == 3]

plt.hist([df_class_1["Age"], df_class_2["Age"], df_class_3["Age"]], bins=20) # rwidth=0.5
plt.legend(["1st Class", "2nd Class", "3rd Class"])
plt.show()
```

```python
plt.scatter(df["Age"], df["Fare"])
plt.xlabel("Age")
plt.ylabel("Ticket Price")
plt.show()
```

```python
df_male = df[df["Sex"] == "male"]
df_female = df[df["Sex"] == "female"]

plt.scatter(df_female["Age"], df_female["Fare"], c="red", alpha=0.2)
plt.scatter(df_male["Age"], df_male["Fare"], c="blue", alpha=0.2)
plt.xlabel("Age")
plt.ylabel("Ticket Price")
plt.legend(["Male", "Female"])
plt.show()
```

```python
df_survived = df[df["Survived"] == 1]
df_died = df[df["Survived"] == 0]

plt.scatter(df_survived["Age"], df_survived["Fare"], c="red", alpha=0.2)
plt.scatter(df_died["Age"], df_died["Fare"], c="blue", alpha=0.2)
plt.xlabel("Age")
plt.ylabel("Ticket Price")
plt.legend(["Survived", "Died"])

plt.savefig("sample.png", format="png")
plt.savefig("sample.jpg", format="jpg")
plt.savefig("sample.tiff", format="tiff")
plt.savefig("sample.svg", format="svg")

plt.show()
```

```python
plt.boxplot(df["Age"].dropna())
plt.show()
```
