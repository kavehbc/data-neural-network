# Titanic / Decision Tree

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
df = pd.read_csv("../titanic.csv")
df.head()
```

```python
df.isna().sum()
```

```python
features = ["Pclass", "Sex", "Age", "SibSp", "Parch", "Embarked"]
label = "Survived"
```

```python
df[features].head()
```

## One-Hot Encoder

```python
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
# encoder.fit(df[["Embarked"]])
encoded = encoder.fit_transform(df[["Sex", "Embarked"]])
encoded_features = encoder.get_feature_names_out()
df_encoded = pd.DataFrame(data=encoded, columns=encoded_features)
df = pd.merge(df, df_encoded, left_index=True, right_index=True)
df.head()
```

```python
features.extend(list(encoded_features))
features.remove("Sex")
features.remove("Embarked")
features
```

```python
df[features].head()
```

## Train / Test Split

```python
X = df[features]
y = df[label]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42)

X_train.shape, X_test.shape
```

## Training

```python
model = DecisionTreeClassifier(max_depth=3, criterion='entropy')
clf = model.fit(X_train, y_train)
clf
```

```python
df[features].sample()
```

```python
fig = plt.figure(figsize=(25,20))
_ = plot_tree(clf, 
               feature_names=features,  
               class_names=["Survived", "Not Survived"],
               filled=True)
```

```python
fig.savefig("decistion_tree.png")
```

## Evaluation

```python
model.score(X_test, y_test)
```

```python
predictions = model.predict(X_test)

cm = confusion_matrix(y_test, predictions, labels=model.classes_)
disp = ConfusionMatrixDisplay(cm, display_labels=model.classes_)
disp.plot(cmap="Greens")
plt.show()
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, predictions)
```

## Feature Importance

```python
feature_importances = clf.feature_importances_
feature_importances
```

```python
series = pd.Series(feature_importances, index=features)

fig, ax = plt.subplots()
series.plot.bar(ax=ax)
ax.set_title("Feature importances using MDI")
ax.set_ylabel("Mean decrease in impurity")
fig.tight_layout()
```
