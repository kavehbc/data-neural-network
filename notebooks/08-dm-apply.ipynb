 # Data Manipulation

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/titanic.csv")
df.head()
```

```python
df["Gender"] = None
df.head()
```

## pd.iterrows()

```python
%%time

for index, row in df.iterrows():
    if row["Sex"] == "male":
        df.loc[index, "Gender"] = 1
        # row["Gender"] = 1
    else:
        df.loc[index, "Gender"] = 0
        # row["Gender"] = 1
```

```python
df.head()
```

```python
for index, row in df.iterrows():
    if index == 0:
        print(row)
        print(type(row))
        print(type(row["Cabin"]))
```

## pd.itertuples()

```python
df["Gender"] = None
```

```python
%%time

for row in df.itertuples():
    if row.Sex == "male":
        df.loc[row.Index, "Gender"] = 1
    else:
        df.loc[row.Index, "Gender"] = 0
```

## np.where()

```python
df["Gender"] = None
```

```python
%%time

df["Gender"] = np.where(df["Sex"] == "male", 1, 0)
```

## pd.apply()

```python
df["Gender"] = None
```

```python
%%time

def gender_encoder(x):
    if x == "male":
        return 1
    else:
        return 0
        
df["Gender"] = df["Sex"].apply(gender_encoder)
```

```python
df.head()
```

```python
# apply on multiple columns

def test(x, y):
    print(x, y)
    print("")

df.apply(lambda x: test(x.Pclass, x.Name), axis=1)
```

```python
# column-wise

def test(x):
    print(x)
    print("")

df.apply(test, axis=0)
```

```python
%%time

import time

for i in trange(10):
    time.sleep(1)
    pass
    
```

```python
%%time

for index, row in tqdm(df.iterrows(), total=df.shape[0], desc="Pandas Processing"):
    time.sleep(0.001)
    pass
```

```python
%%time

with tqdm(total=df.shape[0], bar_format="{percentage:3.0f}%|{bar}{r_bar} {desc:<5}") as pbar:
    for index, row in df.iterrows():
        time.sleep(0.001)
        pbar.update(1)
        # pbar.set_description(row["Name"], refresh=True)
        pbar.desc = row["Name"]
        pass
```

```python
%%time

tqdm.pandas(desc="Progress")

def gender_encoder(x):
    time.sleep(0.001)
    if x == "male":
        return 1
    else:
        return 0
        
df["Gender"] = df["Sex"].progress_apply(gender_encoder)
```

```python
tqdm.pandas(desc="Progress")

pbar = tqdm(total=df.shape[0], bar_format="{percentage:3.0f}%|{bar}{r_bar} {desc:<5}")

def dynamic_desc(x, y):
    time.sleep(0.001)
    pbar.update(1)
    pbar.desc = y
    
df.apply(lambda x: dynamic_desc(x.Pclass, x.Name), axis=1)
```
